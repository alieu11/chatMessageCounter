```python
# note that this was used in Google Collab (Jupyter Notebook equivalent)

import json
import matplotlib.pyplot as plt
import re

# open all the messages

with open('/content/message_1.json') as x:
  m1 = json.loads(x.read())
with open('/content/message_2.json') as x:
  m2 = json.loads(x.read())

# make an array to hold all the messages
allMessages = [m1, m2]
```

```python
def messageCounter():

  # for each participant, count their messages

  """
  NOTE:

  The reason the code is not O(n) is because it chooses the participant first,
  THEN searches the entire list of messages for the respective sender.
  For O(n), the code should parse through the messages,
  and for each sender name, iterate the counter.

  Example:
  message1 has sender1, so add 1 to sender1 counter
  message2 has sender2, so add 1 to sender2 counter
  message3 has sender1, so add 1 to sender1 counter

  This allows for the messages to be read through once,
  instead of (number of participants) times. This would also allow for
  the possibility of a "timelapse".

  TODO:

  Make this program run in O(n)

  """

  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

    # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          count = count + 1
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

    # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = False))
  # print(sortedDict)

  return sortedDict

def printMessageCounter():

  # for each particpant, count their messages (PRINTED BY DESCENDING)

  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

    # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          count = count + 1
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

    # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = True))
  # print(sortedDict)

  print("Total Messages Sent:")

  for keys, values in sortedDict.items():
    print(keys, values)


def totalWordsSpoken():
  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

  # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          if 'content' in k:
            #print(k['content'])
            r = re.findall(r'\w+', k['content'])
            count = count + len(r)
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

  # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = False))
  # print(sortedDict)

  return sortedDict

def printTotalWordsSpoken():
  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

  # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          if 'content' in k:
            #print(k['content'])
            r = re.findall(r'\w+', k['content'])
            count = count + len(r)
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

  # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = True))
  # print(sortedDict)

  print("Total Words Spoken:")

  for keys, values in sortedDict.items():
    print(keys, values)



def totalCharactersSpoken():
  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

  # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          if 'content' in k:
            #print(k['content'])
            r = len(k['content'])
            #print(r)
            count = count + r
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

  # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = False))
  # print(sortedDict)

  return sortedDict




def printTotalCharactersSpoken():
  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

  # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          if 'content' in k:
            #print(k['content'])
            r = len(k['content'])
            count = count + r
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

  # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = True))
  # print(sortedDict)

  print("Total Characters Spoken:")

  for keys, values in sortedDict.items():
    print(keys, values)



def wordCounter(word):

  unsortedDict = dict()
  totalCount = 0

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    count = 0

    # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          if 'content' in k:
            if word in k['content']:
              count = count + 1
              totalCount = totalCount + 1
    #print(i['name'], count)
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary
    tempName = i['name']
    unsortedDict[tempName] = count

    # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1]))
  print("Total: ", totalCount)
  for keys, values in sortedDict.items():
    print(keys, values)

  return sortedDict









def teacupCounter():
  return wordCounter('\u00f0\u009f\u008d\u00b5')

def averageWordsSpoken():
  # initialize empty dictionary
  unsortedDict = dict()

  # load all the participants, initialize count = 0
  for i in m1['participants']:
    messageCount = 0
    wordCount = 0

  # go through all the messages
    for j in allMessages:
      # for each message, if sender name matches the participant name, increment count by 1
      for k in j['messages']:
        if k['sender_name'] == i['name']:
          messageCount = messageCount + 1
          if 'content' in k:
            #print(k['content'])
            r = re.findall(r'\w+', k['content'])
            wordCount = wordCount + len(r)
    #print(i['name'], count)
    # add the name and respective count value to the unsorted dictionary

    tempName = i['name']

    averageWords = modifiedDivison(wordCount, messageCount)

    unsortedDict[tempName] = averageWords

  # sort the dictionary

  # sorts the dictionary by increasing messages
  sortedDict = dict(sorted(unsortedDict.items(), key = lambda item: item[1], reverse = True))
  # print(sortedDict)

  print("Average Words per Message:")

  for keys, values in sortedDict.items():
    print(keys, values)


def printBarGraph(dictData):

  data = dictData
  names = list(data.keys())
  values = list(data.values())

  plt.figure(figsize = (15, 10))
  plt.barh(range(len(data)), values, tick_label = names)
  plt.xticks(rotation = 90)

  plt.show()


def printPieChart(dictData):

  data = dictData
  names = list(data.keys())
  values = list(data.values())

  fig, ax = plt.subplots()
  ax.pie(values, labels = names, radius = 2, autopct = '%1.1f%%')

def modifiedDivison(n, d):
  return n / d if d else 0
```
